```python
# A single insurance customer represented as variables

customer_id = 10042
customer_name = "John Smith"
age = 45
region = "Ontario"
policy_type = "Auto"
annual_premium = 1840.00
num_claims = 2
is_smoker = False
is_active = True

print("Customer:", customer_name)
print("Region:", region)
print("Policy:", policy_type)
print("Annual Premium: $", annual_premium)
print("Claims made:", num_claims)
print("Active policy:", is_active)


```

```python
# adding three more variables to this customer: claim_amount, loss_ratio, and risk_tier (use a string: "Low", "Medium" or "High").
# Calculating loss_ratio from the variables, and assigning risk_tier based on what the loss ratio tells.
claim_amount = [600, 1300]
total_claim_amount = sum(claim_amount)
loss_ratio = total_claim_amount / annual_premium * 100

risk_tier = ""
if loss_ratio >= 80:
    risk_tier = "High"
elif loss_ratio >= 50 and loss_ratio < 80:
    risk_tier = "Medium"
else:
    risk_tier = "Low"

print("Total claimed Amount:", total_claim_amount)
print("Loss ratio:", round(loss_ratio,2), "%")
print("Risk is", risk_tier)
```

```python
# lists and loops

customers = ["John Smith", "Sarah Lee", "Omar Hassan", "Maria Chen"]
premiums = [1840, 950, 2200, 1100]
claims = [1900, 200, 800, 1600]

for i in range(len(customers)):
    ratio = claims[i] / premiums[i] * 100
    if ratio >=80:
        risk_tier = "High"
    elif ratio >=50:
        risk_tier = "Medium"
    else:
        risk_tier = "Low"
    print(customers[i], "| Loss ratio:", round(ratio,1),"% |", risk_tier)
```

```python
# adding more customers 
customers = ["John Smith", "Sarah Lee", "Omar Hassan", "Maria Chen", "Lea Combs", "Jason Brown"]
premiums = [1840, 950, 2200, 1100, 600, 2800]
claims = [1900, 200, 800, 1600, 2000, 0]

for i in range(len(customers)):
    ratio = claims[i] / premiums[i] * 100
    if ratio >=80:
        risk_tier = "High"
    elif ratio >=50:
        risk_tier = "Medium"
    else:
        risk_tier = "Low"
    print(customers[i], "| Loss ratio:", round(ratio,1),"% |", risk_tier)
```

```python
# Portfolio Summary

total_premiums = sum(premiums)
total_claims = sum(claims)
portfolio_loss_ratio = total_claims / total_premiums * 100

high_risk = sum(1 for i in range(len(customers))
                if claims[i] / premiums [i] *100 >=80)

print("\n--- Portfolio Summary ---")
print("Total Premiums: $", total_premiums)
print("Total Claims: $", total_claims)
print("Portfolio loss ratio:", round(portfolio_loss_ratio, 1), "%")
print("High risk customers:", high_risk, "of", len(customers))
```
