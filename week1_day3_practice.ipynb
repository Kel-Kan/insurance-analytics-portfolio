```python
# Pandas
# Medical cost analysis

import pandas as pd

df = pd.read_csv('insurance.csv')

print(df.head())
print(df.shape)
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
# question 1: What is the average charge for smokers vs non-smokers 
print(df.groupby('smoker')['charges'].mean().round(2))
```

```python
# question 2: What is the average charge for male vs females by region
print(df.groupby(['sex','region'])['charges'].mean().round(2).unstack())
```

```python
# question3: average charge for male vs females, smokers vs smokers
result = df.groupby(['sex', 'region', 'smoker'])['charges'].mean().round(2).reset_index()
result.columns = ['Sex', 'Region', 'Smoker','Avg Charge']
print(result)
```

```python
# question 4: bin BMI into standard medical categories
df['bmi_category'] = pd.cut(df['bmi'],
   bins=[0, 18.5, 24.9, 29.9, 100],
   labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

print(df['bmi_category'].value_counts())
```

```python
# question 5: average charge by BMI category and smoking status 
result = df.groupby(['bmi_category','smoker'], observed=True)['charges'].mean().round(2).reset_index()
result.columns = ['BMI Category', 'Smoker', 'Avg Charge']

print(result)

# mumber of smoker vs non-smokers 

print(df['smoker'].value_counts())
print()
print(df['smoker'].value_counts(normalize=True).round(3) * 100)

```

```python
# question 6: percentage of total cost of smokers vs non-smokers 

smoker_charges = df.groupby('smoker')['charges'].sum().round(2)
smoker_charges_pct = (smoker_charges / smoker_charges.sum() * 100).round(1)

print(smoker_charges)
print()
print("% of total cost:")
print(smoker_charges_pct)
```
