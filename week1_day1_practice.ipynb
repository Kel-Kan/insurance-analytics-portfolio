```python
policy_ids=[1001, 1002, 1003, 1004, 1005]
premiums=[1200, 850, 2100, 670, 1540]
claims=[300, 850, 1800, 0, 2200]

total_premiums = sum(premiums)
total_claims = sum(claims)
loss_ratio = total_claims / total_premiums * 100 

print("Total premiums collected:", total_premiums)
print("Total claims paid:", total_claims)
print("Loss ratio:", round(loss_ratio, 2), "%") 
```

```python
for i in range(len(policy_ids)):
    profit = premiums[i] - claims[i]
    print("Policy", policy_ids[i], "- profit/loss:", profit) 
```
